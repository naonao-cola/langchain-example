```python
from langchain.adapters import openai

openai_api_key = "EMPTY"
openai_api_base = "http://localhost:11434/api/generate"
messages = [{"role": "user", "content": "早上好"}]
 
for c in openai.ChatCompletion.create(
        openai_api_key=openai_api_key,
        openai_api_base=openai_api_base,
    messages = messages,
    model="gpt-3.5-turbo",
    temperature=0,
    stream=True
):

    c["choices"][0]['delta'].to_dict_recursive()
```
